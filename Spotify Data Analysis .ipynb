### DATA VISUALIZATION

```python
corr_df = df.select_dtypes(include=['number']) # select numeric column only
# Calculate the correlation matrix
corr_df.corr(method='pearson')
```

```python
# Plot the correlation matrix 
plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(corr_df.corr(), annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="Blues", linewidths=1, linecolor="black")
heatmap.set_title("Correlation Heatmap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
plt.show()
```

There are some significannt correlation: 
- Energy and Loudness has a strong positive correlation (0.8), suggesting that a song becomes louder, it is perceived as more energetics by listeners. 
- Energy and Acousticnes as a strong negative correlation (-0.7), indicating that the more accoustic the song is, the less energetic listerners feel.
- Valence and Danceability has quite strong positive correlation (0.5), indicating that higher danceability tends to be associated with a more positive mood or emotional tone in the music.
